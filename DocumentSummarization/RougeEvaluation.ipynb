```python
from rouge import Rouge
from rouge import FilesRouge
import os
import nltk
import time
import pickle
import io
import math
import numpy as np
import json
```

```python
rouge=Rouge()
path=os.getcwd()
```

```python
def getRougeScore(prefix,threshold, topicnum):
    global rouge
    global path
    f=io.open(os.path.join(path,prefix+'_Summary_'+str(topicnum)+'_Threshold_'+str(threshold)+'.txt'), 'r', encoding='utf-8')
    hypothesis=f.read()
    f=io.open(os.path.join(path+"/Assignement2_IR/GroundTruth",'Topic'+str(topicnum)+'.1'), 'r', encoding='utf-8')
    reference=f.read()
    for i in range(abs(len(reference)-len(hypothesis))):
        if (len(reference)>len(hypothesis)):
            hypothesis+=" "
        else:
            reference+=" "
    scores=rouge.get_scores(reference,hypothesis,avg=True)
    return scores
```

```python
def getFilesRougeScore(prefix,threshold, topicnum):
    global path
    hyp_path=os.path.join(path,prefix+'_Summary_'+str(topicnum)+'_Threshold_'+str(threshold)+'.txt')
    ref_path=os.path.join(path+"/Assignement2_IR/GroundTruth",'Topic'+str(topicnum)+'.1')
    files_rouge = FilesRouge(hyp_path, ref_path)
    scores = files_rouge.get_scores(avg=True)
    return scores
```

```python
prefixes=['DegCentrality','TextRank']
threshold_list=[0.1,0.2,0.3]
topics=range(1,6)
```

```python
for prefix in prefixes[:1]:
    for topicnum in topics:
        for threshold in threshold_list:
            score=getRougeScore(prefix,threshold,topicnum)
            print('Prefix: ',prefix,'. Topic: ',str(topicnum),'. Threshold: ',str(threshold))
            print(score)
            print('\n')
```

```python
for prefix in prefixes[1:]:
    for topicnum in topics:
        for threshold in threshold_list:
            score=getRougeScore(prefix,threshold,topicnum)
            print('Prefix: ',prefix,'. Topic: ',str(topicnum),'. Threshold: ',str(threshold))
            print(score)
            print('\n')
```
